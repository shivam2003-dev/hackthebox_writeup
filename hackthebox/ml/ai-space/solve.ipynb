```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
```

```python
data = np.load('/Users/shivamkumar/Documents/personal/hackthebox/ai_+space/distance_matrix.npy')
print('Shape:', data.shape)
print('Min:', data.min(), 'Max:', data.max())
```

```python
mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42)
X = mds.fit_transform(data)
print('MDS output shape:', X.shape)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(X[:, 0], X[:, 1], s=5, c='cyan', marker='.')
plt.axis('off')
plt.tight_layout()
plt.savefig('/Users/shivamkumar/Documents/personal/hackthebox/ai_+space/flag.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved to flag.png')
```
